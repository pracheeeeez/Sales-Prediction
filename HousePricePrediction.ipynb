Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing Dataset



```python
house_dataset = sklearn.datasets.fetch_california_housing()
print(house_dataset)


```

```python
houseprice_dataframe = pd.DataFrame(house_dataset.data, columns = house_dataset.feature_names)
```

```python
houseprice_dataframe.head()
```

```python
houseprice_dataframe.describe()
```

```python
houseprice_dataframe["Price"] = house_dataset.target
```

```python
houseprice_dataframe.head()
```

```python
print(houseprice_dataframe.shape)
houseprice_dataframe.isnull().sum()

```

```python
correlation = houseprice_dataframe.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar= True, square = True, fmt = '.1f', annot= True, annot_kws= {"size": 8}, cmap= "Reds")
```

```python
X = houseprice_dataframe.drop(["Price"], axis = 1)
Y = houseprice_dataframe["Price"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
model = XGBRegressor()
model.fit(X_train, Y_train)
```

```python
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

Prediction & Evaluation of Training Dataset

```python
score_1 = metrics.r2_score(Y_train,training_data_prediction)
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
print("R squared error :",score_1)
print("Mean Absolute error :",score_2)
```

Prediction and Evaluation of Testing Dataset


```python
testing_data_prediction = model.predict(X_test)
```

```python
score_1 = metrics.r2_score(Y_test, testing_data_prediction)
score_2 = metrics.mean_absolute_error(Y_test, testing_data_prediction)
print("R squared error :",score_1)
print("Mean Absolute error :",score_2)
```

```python
plt.scatter(Y_test, testing_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price v/s Predicted Price")
plt.show()
```
